# **Data Preprocessing**

```
# doing this for categorical features
dummy_fields = ['week', 'store_id', 'sku_id']

for each in dummy_fields:
    dummy = pd.get_dummies(train_data[each], prefix=each, drop_first=False)
    train_data = pd.concat([train_data, dummy], axis=1)

fields_to_drop = ['record_ID','week', 'store_id', 'sku_id']
data = train_data.drop(fields_to_drop, axis=1)

data.head()
```

```
data.info()
```

```
#feature scaling the continuous variables
quant_features = ['total_price', 'base_price', 'units_sold']

scaled_features = {}
for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean)/std

data.head()
```

```
total_rows = len(data)
total_rows
```

```
# splitting the data into training, validation and test sets

valid_size = 0.2
train_size = 0.6

test_data = data[int(-valid_size*total_rows): ]
data = data[ :int(-valid_size*total_rows)]

target_fields = ['units_sold']

features, target = data.drop(target_fields, axis=1), data[target_fields]
test_features, test_target = test_data.drop(target_fields, axis=1), test_data[target_fields]
```

```
train_features, valid_features = features[ :int(train_size*total_rows)], features[int(train_size*total_rows): ]
train_target, valid_target = target[ :int(train_size*total_rows)], target[int(train_size*total_rows): ]
```

```
from torch.utils.data import Dataset, DataLoader
import torch

class SKUDataset(Dataset):

    def __init__(self, features, target, train=True):
        self.features = features
        self.target = target
        self.train = train
    
    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):

        if self.train:
            return torch.tensor(self.features.iloc[idx], dtype=torch.float32), torch.tensor(self.target.iloc[idx], dtype=torch.float32)
        else:
            return torch.tensor(self.features.iloc[idx], dtype=torch.float32)
```

```

train_dataset = SKUDataset(train_features, train_target, train=True)
valid_dataset = SKUDataset(valid_features, valid_target, train=True)
test_dataset = SKUDataset(test_features, test_target, train=True)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)
```

# **Network Architecture**

```
import torch
import torch.nn as nn
```

```
class Network(nn.Module):

    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(238, 32)
        # 1 output node because we are predicting a continuous value
        self.output = nn.Linear(32, 1)

    def forward(self, x):
        x = self.hidden(x)
        x = torch.sigmoid(x)
        x = self.output(x)

        return x
```

```
model = Network()
model
```

```
class RMSLELoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.mse = nn.MSELoss()

    def forward(self, pred, actual):
        return torch.sqrt(self.mse(torch.log(pred + 1), torch.log(actual + 1)))
```

```
import torch.optim as optim

criterion = RMSLELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

gpu_is_available = torch.cuda.is_available()
if gpu_is_available:
    print('Training on GPU!')
    model = model.cuda()
else:
    print('Training on CPU!')
```

```
epochs = 5
valid_loss_min = np.inf

for epoch in range(epochs):
    
    train_loss = 0.
    valid_loss = 0.

    model.train()
    for features, target in train_loader:
        # moving data to GPU, if it's available
        if gpu_is_available:
            features, target = features.cuda(), target.cuda()

        # clearing out the gradients
        optimizer.zero_grad()
        # forward pass
        preds = model(features.float())
        # calculating the loss
        loss = criterion(preds, target.float())
        # backward propagation
        loss.backward()
        # updating the weights
        optimizer.step()

        # updating the loss
        train_loss += loss.item()*features.size(0)

    model.eval()
    for features, target in valid_loader:
        # moving data to GPU, if it's available
        if gpu_is_available:
            features, target = features.cuda(), target.cuda()
        # forward pass
        preds = model(features.float())
        # calculating the loss
        loss = criterion(preds, target.float())

        valid_loss += loss.item()*features.size(0)

    train_loss /= len(train_loader.dataset)
    valid_loss /= len(valid_loader.dataset)

    print('Epoch #{}, training loss : {:.6f}, validation loss : {:.6f}'
    .format(epoch+1, train_loss, valid_loss))

    if valid_loss < valid_loss_min :
        print('Validation loss decreased from {:.6f} to {:.6f}. Saving model...'
        .format(valid_loss_min, valid_loss))
        valid_loss_min = valid_loss
        torch.save(model.state_dict(), 'model.pt')
```

```
# evaluating performance on the test set
model.load_state_dict(torch.load('model.pt'))
print('Best model loaded from memory!')

test_loss = 0.
model.eval()
for i, (features, target) in enumerate(test_loader):
    # checking if GPU is available
    if gpu_is_available:
        features, target = features.cuda(), target.cuda()
        model = model.cuda()
    
    # forward pass
    preds = model(features)
    # calculating loss
    loss = criterion(preds, target)
    test_loss += loss.item()*features.size(0)

    print('{} batch(es) done..'.format(i+1))

test_loss /= len(test_loader.dataset)
print('Testing loss : {:.6f}'.format(test_loss))
```
