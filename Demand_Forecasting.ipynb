# **Data Preprocessing**

```python
def get_dummy_fields(data): 
    
    # doing this for categorical features
    dummy_fields = ['week', 'store_id', 'sku_id']

    for each in dummy_fields:
        dummy = pd.get_dummies(data[each], prefix=each, drop_first=False)
        data = pd.concat([data, dummy], axis=1)

    return data
```

```python
def drop_fields(data):
    
    fields_to_drop = ['record_ID','week', 'store_id', 'sku_id']
    data = data.drop(fields_to_drop, axis=1)
    return data
```

```python
 def scale_features(data, train=True):
        
    #feature scaling the continuous variables
    quant_features = ['total_price', 'base_price']
    
    if train:
        quant_features.append('units_sold')

    scaled_features = {}
    for each in quant_features:
        mean, std = data[each].mean(), data[each].std()
        scaled_features[each] = [mean, std]
        data.loc[:, each] = (data[each] - mean)/std

    return data, scaled_features
```

```python
from torch.utils.data import Dataset, DataLoader
import torch

class SKUDataset(Dataset):

    def __init__(self, features, target=None, train=True):
        self.features = features
        self.target = target
        self.train = train
    
    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):

        if self.train:
            return torch.tensor(self.features.iloc[idx], dtype=torch.float32), torch.tensor(self.target.iloc[idx], dtype=torch.float32)
        else:
            return torch.tensor(self.features.iloc[idx], dtype=torch.float32)
```

```python

train_dataset = SKUDataset(train_features, train_target, train=True)
valid_dataset = SKUDataset(valid_features, valid_target, train=True)
test_dataset = SKUDataset(test_data, train=False)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)
```

# **Network Architecture**

```python
import torch
import torch.nn as nn
```

```python
class Network(nn.Module):

    def __init__(self):
        super().__init__()
        self.hidden1 = nn.Linear(250, 64)
        self.hidden2 = nn.Linear(64, 16)
        # 1 output node because we are predicting a continuous value
        self.output = nn.Linear(16, 1)

    def forward(self, x):
        x = self.hidden1(x)
        x = torch.sigmoid(x)
        x = self.hidden2(x)
        x = torch.sigmoid(x)
        x = self.output(x)

        return x
```

```python
model = Network()
model
```

```python
class RMSLELoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.mse = nn.MSELoss()

    def forward(self, pred, actual):
        return torch.sqrt(self.mse(torch.log(pred + 1), torch.log(actual + 1)))
```

```python
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

gpu_is_available = torch.cuda.is_available()
if gpu_is_available:
    print('Training on GPU!')
    model = model.cuda()
else:
    print('Training on CPU!')
```

```python
epochs = 30
valid_loss_min = np.inf

for epoch in range(epochs):
    
    train_loss = 0.
    valid_loss = 0.

    model.train()
    for features, target in train_loader:
        # print('features :{}, target: {}'.format(features, target))
        # moving data to GPU, if it's available
        if gpu_is_available:
            features, target = features.cuda(), target.cuda()

        # clearing out the gradients
        optimizer.zero_grad()
        # forward pass
        preds = model(features)
        # calculating the loss
        loss = criterion(preds, target)
        # backward propagation
        loss.backward()
        # updating the weights
        optimizer.step()

        # updating the loss
        train_loss += loss.item()*features.size(0)
        

    model.eval()
    for features, target in valid_loader:
        # moving data to GPU, if it's available
        if gpu_is_available:
            features, target = features.cuda(), target.cuda()
        # forward pass
        preds = model(features)
        # calculating the loss
        loss = criterion(preds, target)

        valid_loss += loss.item()*features.size(0)

    train_loss /= len(train_loader.dataset)
    valid_loss /= len(valid_loader.dataset)

    print('Epoch #{}, training loss : {:.6f}, validation loss : {:.6f}'
    .format(epoch+1, train_loss, valid_loss))

    if valid_loss < valid_loss_min :
        print('Validation loss decreased from {:.6f} to {:.6f}. Saving model...'
        .format(valid_loss_min, valid_loss))
        valid_loss_min = valid_loss
        torch.save(model.state_dict(), 'model.pt')
```
